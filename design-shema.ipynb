```python
import pandas as pd
import numpy as np
```

```python
champs = pd.read_csv("champs.csv")
matches = pd.read_csv("clean_matches.csv")
participants = pd.read_csv("clean_participants.csv")
playerstats = pd.read_csv("clean_player_stats.csv")
teambans = pd.read_csv("clean_teambans.csv")
teamstats = pd.read_csv("clean_teamstats.csv")
```

```python
participants_new = participants.merge(champs, how='left', left_on='championid', right_on='id')
participants['championid'] = participants_new['name']
participants.rename(columns={'championid': 'champion'}, inplace=True)

participants.head()
```

```python
teambans_new = teambans.merge(champs, how='left', left_on='championid', right_on='id')
teambans['championid'] = teambans_new['name']
teambans.rename(columns={'championid': 'champion'}, inplace=True)

teambans.head()
```
